```python
import pandas as pd
from matplotlib import pyplot
```

```python
df = pd.read_excel(r'C:\Users\18143\Desktop\COVID-19 Transmission.xlsx')
print(df)
```

```python
stats = df.describe(include='all')
print(stats)
```

# statistical methods to find corelationship between variables. correlation variable lies between +1 to -1. Positive sign indicates the positive relationship between two variables. value goes towards the 0 indicate the weak relationship between two variables. Negative sign indicate the inverse relationship between two variables. 

Three different stats method to calculate corelation coefficient are pearson, kendall, spearman.


```python
df.corr(method='pearson')
```

```python
df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```

```python
ax = df.hist(column='Total Overseas acquired', bins=25, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)

ax = ax[0]
for x in ax:

    # Despine
    x.spines['right'].set_visible(False)
    x.spines['top'].set_visible(False)
    x.spines['left'].set_visible(False)

    # Switch off ticks
    x.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

    # Draw horizontal axis lines
    vals = x.get_yticks()
    for tick in vals:
        x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1)

    # Remove title
    x.set_title("")

    # Set x-axis label
    x.set_xlabel("COVID-19 cases_numbers ", labelpad=20, weight='bold', size=12)

    # Set y-axis label
    x.set_ylabel("COVID-19 cases", labelpad=20, weight='bold', size=12)
```
